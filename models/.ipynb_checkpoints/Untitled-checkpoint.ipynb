## Model training

```python
x1 = df.drop('Item_Outlet_Sales',axis=1)
y1 = df['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1,y1, test_size=0.2,random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Create instance
from sklearn.linear_model import LinearRegression
model = LinearRegression()
#model fit
model.fit(x_train, y_train)
```

```python
# Prediction of model
y_pred_train = model.predict(x_train)
y_pred_train
```

```python
# Prediction of model
y_pred_test = model.predict(x_test)
y_pred_test
```

## Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## On Training data

```python
residual_train = (y_train - y_pred_train)

mse = mean_squared_error(y_train,y_pred_train)
print("mean Square Error: ",mse)

mae = mean_absolute_error(y_train,y_pred_train)
print("mean absolute error: ",mae )

rmse = np.sqrt(mse)
print("root mean square error: ",rmse )

r2 = r2_score(y_train,y_pred_train)
print("r square: ",r2 )
```

## On Testing data

```python
residual_test = (y_test - y_pred_test)

mse = mean_squared_error(y_test, y_pred_test)
print("mean Square Error: ",mse)

mae = mean_absolute_error(y_test,y_pred_test)
print("mean absolute error: ",mae )

rmse = np.sqrt(mse)
print("root mean square error: ",rmse )

r2 = r2_score(y_test,y_pred_test)
print("r square: ",r2 )
```

## construction of boxplot for all columns

```python
def diagnostic_plots(df):
    # function takes a dataframe (df) and
    # the list of variables of interest as arguments

    col=df._get_numeric_data().columns
    print(col)
    
    for ele in col:
        
        # define figure size
        plt.figure(figsize=(4, 2))
        
        # boxplot
       
        sns.boxplot(y=df[ele])
        plt.title('Boxplot')
        
        
        plt.show()
diagnostic_plots(x1)
```
